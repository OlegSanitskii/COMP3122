## 2. Designing Baseline

One of the most practical evaluation tools is a strong `baseline`â€”a simple method that gives you something to beat.

Good baselines help you measure progress, catch bugs early, and understand the difficulty of your task.

Sometimes they can even be surprisingly hard to beat.

<img src="https://raw.githubusercontent.com/PyDataGBC/PyData2025/refs/heads/main/Lab_Week11_LogisticRegression/assets/baseline_for_classification.png" width="600" height="500" />

#### Majority class baseline: What accuracy score would you expect from the optimistic classifier that expects no serious trouble at all?
 - This is our `dummy` classifier that predicts 0 for all samples
 - This is called the `Null accuracy`
 - Verify that accuracy score using the `accuracy_score()` function


```python
#your code here
y_true = df["SeriousDlqin2yrs"]
y_pred = np.zeros_like(y_true)

accuracy_score(y_true, y_pred) * 100
```

#### Accuracy is not the only metric to evaluate a classifier!
#### There are other important metrics such as `precision`, `recall` and `F1` score (you will learn about them in the future sections of this lab)
#### Calculate the precision, recall and F1 score of the optimistic classifier.
- Use the `precision_score()`, `recall_score()` and `f1_score()` functions from `sklearn.metrics` module and pass `y_pred` and `y_true` values to them.

```python
#your code here
precision_null = precision_score(y_true, y_pred, zero_division=0)
recall_null = recall_score(y_true, y_pred, zero_division=0)
f1_null = f1_score(y_true, y_pred, zero_division=0)

print("Null precision:", precision_null)
print("Null recall   :", recall_null)
print("Null F1       :", f1_null)
```

# 4. Logistic Regression Model

#### Train a LogesticRegression model using ALL the data and find the accuracy_score
#### Set `max_iter=1000` in the LogisticRegression constructor to avoid a warning message


#### Call this model as `model`

Hint: There is no need to split the data

```python
# your code here
model = LogisticRegression(max_iter=1000)

# fit your model with X and y
model.fit(X, y)

# predict the values and store them in y_pred_base
y_pred_base = model.predict(X)
```

#### How many people in the prediction has serious trouble with their loan payment (`Label 1`)?

```python
#your code here
np.sum(y_pred_base == 1)
```

#### How much better was the LogisticRegression model compared to the Null in terms of accuracy?

```python
y_true = df["SeriousDlqin2yrs"]
y_pred_null = np.zeros_like(y_true)

null_acc = accuracy_score(y_true, y_pred_null)
print("Null accuracy:", null_acc * 100, "%")


base_acc = accuracy_score(y, y_pred_base)

improvement = (base_acc - null_acc) * 100

print(f"Accuracy of base model: {base_acc * 100:.4f} %")
print(f"Accuracy of null model: {null_acc * 100:.4f} %")
print(f"Difference (base - null): {improvement:.4f} percentage points")
```

# 5. Dealing with Imbalanced Data: Assigning Higher Weights to the Minority Class

#### Running the logistic regression model with different `class_weight`

#### For class 0 (no trouble) set the weight to 1 and for class 1 (with trouble) set the weight to 2 - we give higher weight to the minority class


#### Call this model as `model_weighted`


```python
#your code here
model_weighted = LogisticRegression(max_iter=1000, class_weight={0: 1, 1: 2})

# (1)fit your model and (2)predict the value and (3)compute accuracy
model_weighted.fit(X, y)
y_pred_weighted = model_weighted.predict(X)

accuracy_score(y, y_pred_weighted)
```

#### How many people in the prediction made by `model_weighted` has serious trouble with their loan payment (`Label 1`)?

```python
np.sum(y_pred_weighted == 1)
```

#### Evaluate the weighted model using the confusion_matrix, precision, recall and F1 score

```python
#your code here
confusion_matrix(y, y_pred_weighted)
```

```python
precision_weighted = precision_score(y, y_pred_weighted)
recall_weighted = recall_score(y, y_pred_weighted)
f1_weighted = f1_score(y, y_pred_weighted)

print("Weighted precision:", precision_weighted)
print("Weighted recall   :", recall_weighted)
print("Weighted F1       :", f1_weighted)
```

#### Compare and Contrast the results of the `model_weighted` with the `model`

#### Repeat the above with different values of `w` in the range between 1 and 10. Plot the accuracy score as a function of `w` label 1's weight


```python
#your code here
weights = range(1, 11)
accuracies = []

for w in weights:
    model_tmp = LogisticRegression(max_iter=1000,
                                   class_weight={0: 1, 1: w})
    model_tmp.fit(X, y)
    y_pred_tmp = model_tmp.predict(X)
    acc = accuracy_score(y, y_pred_tmp)
    accuracies.append(acc)

plt.plot(list(weights), accuracies, marker="o")
plt.xlabel("Weight for class 1")
plt.ylabel("Accuracy")
plt.title("Accuracy vs class 1 weight")
plt.show()

accuracies
```

#### Based on the plot of scores, what is the best value for label 1 weight?

The accuracy increases slightly when the class 1 weight goes from 1 to 2, reaching its maximum at weight = 2. For weights larger than 2, the accuracy continuously decreases, meaning the model becomes worse. Therefore, the optimal weight for class 1 in terms of accuracy is **2**.

#### Scikit Learn also has a function called `compute_class_weight` that automatically computes the weights based on the class distribution.

```python
from sklearn.utils.class_weight import compute_class_weight
classes = np.array([0, 1])
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y)
class_weights
```

#### Make a logestic regression model with the computed class weights and evaluate it using the confusion_matrix, precision, recall and F1 score


#### Call this model as `auto_weighted model`

```python
auto_weights = {0: class_weights[0], 1: class_weights[1]}
auto_weighted_model = LogisticRegression(max_iter=1000,
                                         class_weight=auto_weights)

auto_weighted_model.fit(X, y)
y_pred_auto = auto_weighted_model.predict(X)

print("Confusion matrix (auto weights):")
print(confusion_matrix(y, y_pred_auto))

precision_auto = precision_score(y, y_pred_auto)
recall_auto = recall_score(y, y_pred_auto)
f1_auto = f1_score(y, y_pred_auto)

print("Auto-weighted precision:", precision_auto)
print("Auto-weighted recall   :", recall_auto)
print("Auto-weighted F1       :", f1_auto)
```

#### Compare and Contrast the results of the `auto_weighted model` with the `weighted model`

#### Which model has higher type 1 error? and which model has higher type 2 error?

**Type 1 error (False Positives):**
- Weighted model FP = 1,921  
- Auto-weighted FP = 7,724  

The **auto_weighted model has much higher Type 1 error**, because it predicts many more customers as â€œin troubleâ€ even when they are not.

**Type 2 error (False Negatives):**
- Weighted model FN = 6,119  
- Auto-weighted FN = 3,545  

The **weighted model has higher Type 2 error**, because it misses more customers who are actually in serious trouble.

**Summary:**
- `auto_weighted_model` â†’ higher Type 1 error (more false alarms)  
- `weighted_model` â†’ higher Type 2 error (misses more risky customers)


#### If we want to give people very large amount of credit, which model should we use? and why?

If we want to give people **very large amounts of credit**, it is much more important to avoid lending money to customers who are actually highâ€“risk. In other words, we want to **minimize Type 2 error (FN)** â€” cases where the model predicts â€œno troubleâ€ but the person is in serious trouble.

Among our models, the **auto_weighted_model** has:

- much **lower FN** and
- much **higher recall for class 1**

than the weighted model, even though it has more false positives.

Therefore, for very large credit limits, we should choose the **auto_weighted_model**, because it is more conservative: it catches more truly risky customers and reduces the chance of giving a big loan to someone who is likely to default, even at the cost of more false alarms.
